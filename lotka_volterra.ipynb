```
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 15), dpi=80, facecolor='w', edgecolor='k')
```

```

c0_prey = 100
c0_pred = 5
beta = 0.0015
Y = .35
sigma = .09
mu = 0.048
M = 1000

num_iter = 8000
```

```
c_prey = c0_prey
c_pred = c0_pred

c_pred_list = []
c_prey_list = []

for i in range(num_iter):

  c_prey_new = c_prey + (mu * ((M - c_prey) / M) *c_prey) - (beta * c_prey * c_pred)
  #c_prey_new = c_prey + (mu *c_prey) - (beta * c_prey * c_pred)
  c_pred_new = c_pred + (Y * beta * c_prey * c_pred) - (sigma * c_pred)

  c_prey, c_pred = c_prey_new, c_pred_new

  c_prey_list.append(c_prey)
  c_pred_list.append(c_pred)
  
```

```
plt.figure(figsize=(20, 9))
plt.plot(c_prey_list)
plt.plot(c_pred_list)
plt.show()
```

```
plt.plot(c_prey_list, c_pred_list)
plt.show()
```
